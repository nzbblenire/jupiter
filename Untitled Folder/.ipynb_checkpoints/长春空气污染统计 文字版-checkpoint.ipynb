```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import re
import numpy as np

def get_air(htmlweb):
    html = urlopen(htmlweb)
    air = bs(html,"html.parser")
    list = []
    for i in air.findAll("td",{"class" : re.compile("aqi-lv[0-9]")}):
        list.append(i.get_text())
    list_str_dealing(list)
    return list

def list_str_dealing(list):
    for i in range(len(list)):
        if list[i] == '\r\n                                       严重污染':
            list[i] = "严重污染"
        if list[i] == '\r\n                                       重度污染':
            list[i] = "重度污染"
        if list[i] == '\r\n                                       中度污染':
            list[i] = "中度污染"
        if list[i] == '\r\n                                       轻度污染':
            list[i] = "轻度污染"
        if list[i] == '\r\n                                       良':
            list[i] ="良"
        if list[i] == '\r\n                                       优':
            list[i] ="优"
    return list

def group_air(list):
    airA = 0
    airB = 0
    airC = 0
    airD = 0
    airE = 0
    airF = 0
    for i in range(len(list)):
        if list[i] == "严重污染" :
            airF += 1
        elif list[i] == "重度污染":
            airE += 1
        elif list[i] == "中度污染":
            airD += 1
        elif list[i] == "轻度污染":
            airC += 1
        elif list[i] == "良":
            airB += 1
        elif list[i] == "优":
            airA += 1
    list = [airA,airB,airC,airD,airE,airF]
    return list

air_month_array = np.array([0, 0, 0, 0, 0, 0])
for i in range(12):
    if i < 10 and i >0 :
        list = get_air("http://www.tianqihoubao.com/aqi/changchun-20170"+str(i)+".html")
        print ("第"+str(i)+"月数据:"+str(list))
        list_gourp = []
        list_group = group_air(list)
        for i in range(6):
            print ("level"+str(i)+": "+str(list_group[i])+"days")
        air_month_part = np.array(list_group)
        air_month_current = np.array([air_month_part])
        print ("该月:level"+str(air_month_current.argmax())+"天数最多")
        print ("该月:level"+str(air_month_current.argmin())+"天数最少")
        air_month_array += air_month_part
            
    elif i >=10 :
        list = get_air("http://www.tianqihoubao.com/aqi/changchun-2017"+str(i)+".html")
        print ("第"+str(i)+"月数据:"+str(list))
        list_gourp = []
        list_group = group_air(list)
        for i in range(6):
            print ("level"+str(i)+": "+str(list_group[i])+"days")
        air_month_part = np.array(list_group)
        air_month_current = np.array([air_month_part])
        print ("该月:level"+str(air_month_current.argmax())+"天数最多")
        print ("该月:level"+str(air_month_current.argmin())+"天数最少")
        air_month_array += air_month_part
print (air_month_array)
print ("most days")
print (air_month_array.argmax())
print ("least days")
print (air_month_array.argmin())  #attention to using str()
#list = get_air("http://www.tianqihoubao.com/aqi/changchun-201711.html")
#print (list)
```

```python
def change_xy_axis(list):
    new_list = []
    for j in range(len(list[0])):
        new_list.append([])
        for i in range(len(list)):
            new_list[j].append(list[i][j])
    return new_list
#print (change_xy_axis(air_year))
air_year_df = pd.DataFrame(change_xy_axis(air_year),
                           index = ["level0","level1","level2","level3","level4","level5"])
print (air_year_df)
```
